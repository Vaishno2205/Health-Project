```python
import pandas as pd

file_path = '/content/drive/MyDrive/lung_cancerP.csv'  # your actual path
df = pd.read_csv(file_path)

df.head()
```

```python
# Identify binary categorical columns
binary_cols = ['GENDER', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE',
               'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING', 'ALCOHOL_CONSUMING',
               'COUGHING', 'SHORTNESS_OF_BREATH', 'SWALLOWING_DIFFICULTY', 'CHEST_PAIN',
               'LUNG_CANCER']

# Create a copy of the DataFrame to avoid modifying the original
df_numeric = df.copy()

# Convert binary categorical columns to numeric (1s and 0s)
# Assuming 'Yes'/'M' maps to 1 and 'No'/'F' maps to 0
mapping = {'Yes': 1, 'No': 0, 'M': 1, 'F': 0}
for col in binary_cols:
    df_numeric[col] = df_numeric[col].map(mapping)

# Display the first few rows of the new DataFrame to verify the conversion
display(df_numeric.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix for the numeric DataFrame
correlation_matrix_numeric = df_numeric.corr()

# Create the heatmap
plt.figure(figsize=(14, 10)) # Adjust figure size for better readability
sns.heatmap(correlation_matrix_numeric, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

# Display the heatmap
plt.title('Correlation Heatmap (including converted categorical columns)')
plt.show()
```
