```python
import pandas as pd

file_path = '/content/drive/MyDrive/heart_p.csv'  # your actual path
df = pd.read_csv(file_path)

df.head()
```

```python
# prompt: Using dataframe df: create the heatmap with the values(include the values in the heatmap, without the overlap of values)

import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of df')
plt.show()
```
